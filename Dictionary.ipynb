```python
import requests


class Definition:

    @staticmethod
    def get_definition(word):
        """Fetch definitions for a word from the Dictionary API."""
        url = f"https://api.dictionaryapi.dev/api/v2/entries/en/{word}"
        try:
            response = requests.get(url, timeout=5)  # Timeout for network issues
            response.raise_for_status()
            def_info = response.json()
        except requests.exceptions.ConnectionError:
            return None, "connection_error"
        except requests.exceptions.Timeout:
            return None, "timeout"
        except (ValueError, Exception):
            return None, "not_found"

        try:
            definitions = def_info[0]["meanings"]
        except (KeyError, IndexError):
            return None, "not_found"

        ans = []
        for p in definitions:
            p_speech = p["partOfSpeech"]
            for d in p["definitions"]:
                definition = d["definition"]
                example = d.get("example", None)  # Include usage example if available
                if example:
                    ans.append(f"{p_speech}: {definition}\n   Example: {example}")
                else:
                    ans.append(f"{p_speech}: {definition}")

        synonyms = def_info[0].get("meanings", [{}])[0].get("synonyms", [])
        if synonyms:
            ans.append(f"Synonyms: {', '.join(synonyms[:5])}")  # Show up to 5 synonyms

        return ans, None


def main():
    print("Welcome to the Dictionary App!")
    print()

    while True:
        option = input("Press 1 to look up a word. Press 2 to exit: ").strip()
        print()

        if option not in ["1", "2"]:
            print("Please enter a valid option (1 or 2).")
            print()
            continue

        if option == "1":
            user_input = input("Enter your word: ").strip().lower()  # Normalise input
            print()

            if not user_input:
                print("Please enter a word.")
                print()
                continue

            definitions, error = Definition.get_definition(user_input)

            if error == "connection_error":
                print("No internet connection. Please check your network and try again.")
            elif error == "timeout":
                print("The request timed out. Please try again.")
            elif error == "not_found" or not definitions:
                print(f"No definitions found for '{user_input}'. Check the spelling and try again.")
            else:
                print(f"Definitions for '{user_input}':")
                print()
                for d in definitions:
                    print("*", d)
            print()

        elif option == "2":
            print("Thank you for using the Dictionary App!")
            break


if __name__ == "__main__":
    main()







```
